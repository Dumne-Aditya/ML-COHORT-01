```python
import matplotlib.pyplot as plt  # visual
import seaborn as sns  # visual adv
import pandas as pd  # pre-processing
import numpy as np  # mathmatical array
df_large = pd.read_excel(r"C:\Users\udaya\Downloads\large_sales_data.xlsx")
df_large.info()
df_large.isnull().sum()
df_large.drop_duplicates()
df_large.head(2)
df_large.columns
df_large.nunique()
df_large['Product'].unique()
df_large['Product'].unique()
```

```python
# 1. Bar Plot (alternative to Pie Chart): Sales by Region
```

```python
# 1. Bar Plot (alternative to Pie Chart): Sales by Region
region_sales = df_large.groupby('Region')['Sales'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=region_sales, x='Region', y='Sales', palette='pastel')
plt.title('Sales Distribution by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
print('-------------')
```

```python
# 2. Histogram: Distribution of Sales
plt.figure(figsize=(10, 6))
sns.histplot(df_large['Sales'], bins=20, kde=False, color='salmon', edgecolor='black')
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
print('-------------')
```

```python
# 3. Line Chart: Monthly Sales Trend
month_sales = df_large.groupby('Month')['Sales'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.lineplot(data=month_sales, x='Month', y='Sales', marker='o', color='green')
plt.title('Sales Trend Over Months')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.tight_layout()
plt.show()
print('-------------')
```

```python
region_sales = df_large.groupby('Region')['Sales'].sum().reset_index()
# 4. Scatter Plot: Sales vs Region
plt.figure(figsize=(10, 6))
sns.scatterplot(data=region_sales, x='Region', y='Sales', color='purple')
plt.title('Sales vs Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
print('-------------')
```

```python
# 5. Box Plot: Sales by Region (detecting outliers)
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_large, x='Region', y='Sales')
plt.title('Sales Distribution by Region (Box Plot)')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
print('-------------')
```

```python
# 6. Violin Plot: Sales distribution shape by Region
plt.figure(figsize=(10, 6))
sns.violinplot(data=df_large, x='Region', y='Sales')
plt.title('Sales Distribution by Region (Violin Plot)')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
print('-------------')
```
